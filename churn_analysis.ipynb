```python
df['TotalCharges']=df['TotalCharges'].notnull()
```

```python
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df.drop(columns='customerID',inplace=True,axis=1)
```

```python
df_dummy = pd.get_dummies(df.drop(columns=['Churn','tenure','MonthlyCharges','TotalCharges'],axis=1))
```

```python
df_dummy.shape
```

```python
df_ = pd.concat([df_dummy,df['tenure'],df['MonthlyCharges'],df['TotalCharges']],axis=1)
```

```python
df_.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
sc.fit(df_)
df_feat = sc.transform(df_)
```

```python
df_feature = pd.DataFrame(df_feat,columns=df_.columns[:])
```

```python
df_feature.head()
```

```python
df_feature.isna().sum()
```

```python
df_feature.shape
```

```python
df['Churn'].shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_feature,df['Churn'],test_size=0.25,random_state=0)
```

```python
print("train shape",X_train.shape,y_train.shape)
print("test shape",X_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_1 = knn.predict(X_test)
    error_rate.append(np.mean(pred_1 != y_test))
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,40),error_rate,color='red',linestyle='dashed',marker='o')
plt.title('error_rate vs k_value')
plt.xlabel("error-rate")
plt.ylabel("k value")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=22)
knn.fit(X_train,y_train)
pred_11 = knn.predict(X_test)
```

```python
print(classification_report(y_test,pred_11))
```

```python
print(confusion_matrix(y_test,pred_11))
```

```python
from sklearn.metrics import roc_auc_score , roc_curve
```

```python
y_test_ = knn.predict_proba(X_test)
print("test data {} ".format(roc_auc_score(y_test,y_test_[:,-1])))
y_train_= knn.predict_proba(X_train)
print("train data {} ".format(roc_auc_score(y_train,y_train_[:,-1])))
```

```python
y_test_p = y_test_[:,-1]
y_train_pt =y_train_[:,-1]
```

```python
fpr_test , tpr_test , thresholds = roc_curve(y_test,y_test_p)
fpr_train , tpr_train ,thresholds = roc_curve(y_train,y_train_pt)
```

```python
plt.figure(figsize=(7,7))
plt.plot(fpr_test,tpr_test,color='orange',label='roc_curve_test',linestyle='dashed')
plt.plot(fpr_train,tpr_train,color='magenta',label='roc_curve_train',linestyle='dashed')
plt.plot([0,1],[0,1])
plt.title('roc_curve')
plt.xlabel('true-positive rate')
plt.ylabel('false-positive rate')
plt.legend()
plt.show
```
